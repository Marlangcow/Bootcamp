## 데이터 분석

```python
# 이용금액과 이용자수가 많은 업종과 지역 Top 10 뽑기

jeju_card_region.sort_values(by=['이용금액'], ascending=False)
```

```python
jeju_card_region.groupby(['업종명', '읍면동명']).sum(numeric_only=True).sort_values(by=['이용금액'], ascending=False).head(10)
```

```python
# 인당 이용금액이 많은 지역 Top 10 & 각 지역에서 어떤 업종들이 활성화되어 있는지

jeju_card_region['인당 이용금액'] = jeju_card_region['이용금액']/jeju_card_region['이용자수']
jeju_card_region.head()
```

```python
jeju_card_region.groupby(['인당 이용금액', '업종명']).sum(numeric_only=True).sort_values(by=['인당 이용금액'], ascending=False).head(10)
```

```python
jeju_card_region.groupby(['읍면동명', '업종명']).sum(numeric_only=True).sort_values(by=['이용금액'], ascending=False).head(10)
```

```python
# 카페 이용자수와 유동인구의 관계

jeju_card_region['업종명'].unique()
```

```python
jeju_cafe = jeju_card_region.loc[jeju_card_region['업종명'] == '비알콜 음료점업']
```

```python
sns.barplot(data=jeju_cafe, x='읍면동명', y='이용자수')
plt.rcParams['figure.figsize'] = (100, 5)
plt.rc('font', family='AppleGothic')
```

```python
# 연월일 문자열로 변경 후, datetime으로 변환('0000-00-00'

jeju_population['연월일'] = jeju_population['연월일'].astype('str')
jeju_population.info()
```

```python
jeju_population['연월일'] = pd.to_datetime(jeju_population['연월일'])
jeju_population.rename(columns={'연월일':'연월'}, inplace=True)
```

```python
jeju_population.replace('남', '남성', inplace=True)
jeju_population.replace('여', '여성', inplace=True)
jeju_population
```

```python
jeju_card_region['연월'].astype('str')
jeju_card_region['연월'] = pd.to_datetime(jeju_card_region['연월'])
jeju_card_region.info()
```

```python
jeju_card_region
```

```python
jeju_total = pd.merge(jeju_population, jeju_card_region, how='outer')
jeju_cafe = jeju_total.loc[jeju_total['업종명'] == '비알콜 음료점업']
jeju_total
```

```python
jeju_cafe_corr = jeju_cafe.groupby(['읍면동명', '성별', '연령대']).sum(numeric_only=True)
jeju_cafe_corr
```

```python
jeju_cafe_corr.corr()
```

```python
plt.figure(figsize=(5,5))
plt.rc('font', family='AppleGothic')
sns.heatmap(data=jeju_cafe_corr.corr(), annot=True)
plt.title('카페 이용자수와 유동인구의 관계')
```
